```python
#using box plot
top_categories = df['Categories'].value_counts().nlargest(10).index
plt.figure(figsize=(10, 5))
sns.countplot(x='Categories', data=df, order=top_categories)
plt.xticks(rotation=45)
plt.title("Top 10 content Categories amang streamers")
plt.xlabel("Categories")
plt.ylabel("Number of Streamers")
plt.show()
```

```python
#Calculating the corr of the matrix.
matrix_correlation = df[['Subscribers','Visits','Likes','Comments']].corr()
matrix_correlation
```

```python
#scatter plot of correlation between Subscribers and Likes
def formatter(x,pos):
    return f'{x/1e6:.0f}M'
plt.figure(figsize=(10,5))
xy =sns.scatterplot(x="Subscribers", y ="Likes", data=df)
xy.xaxis.set_major_formatter(FuncFormatter(formatter))
xy.yaxis.set_major_formatter(FuncFormatter(formatter))
plt.title("The correlation between Subscribers and Likes")
plt.xlabel("Subscribers")
plt.ylabel("Likes")
plt.show()
```

```python
#scatter plot of correlation between Subscribers and Visits
plt.figure(figsize=(10,5))
xy =sns.scatterplot(x="Subscribers", y ="Visits", data=df)
xy.xaxis.set_major_formatter(FuncFormatter(formatter))
xy.yaxis.set_major_formatter(FuncFormatter(formatter))
plt.title("The correlation between Subscribers and Visits")
plt.xlabel("Subscribers")
plt.ylabel("Visits")
plt.show()
```

```python
#scatter plot of correlation between Subscribers and Comments
plt.figure(figsize=(10,5))
xy =sns.scatterplot(x="Subscribers", y ="Comments", data=df)
xy.xaxis.set_major_formatter(FuncFormatter(formatter))
plt.title("The correlation between Subscribers and Comments")
plt.xlabel("Subscribers")
plt.ylabel("Comments")
plt.show()
```

```python
#plotting the d0istribution of streams by each country with subscribers.
country_subs = df.groupby(['Country']).sum().reset_index()
country_subs = country_subs.sort_values(by="Subscribers", ascending=False)

def formatter(x, pos):
    return f'{x/1e6}M'

plt.figure(figsize=(10,5))
xy = sns.barplot(x='Country', y='Subscribers', data=country_subs)
xy.yaxis.set_major_formatter(FuncFormatter(formatter))

plt.title("Distribution of streamers in each Country ")
plt.xlabel("Country")
plt.ylabel("Number of Streamers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

country_subs = df.groupby(['Country']).sum().reset_index()
country_subs = country_subs.sort_values(by="Subscribers", ascending=False)

top_ten_countries = country_subs['Country'].head(10)
# DataFrame for top 10 countries
top_ten_countries_df = df[df['Country'].isin(top_ten_countries)]

plt.figure(figsize=(10,5))
sns.countplot(x='Country', data=top_ten_countries_df, palette='viridis', order=top_ten_countries)
plt.title("Subscribers by Country (Top 10)")
plt.xlabel("Country")
plt.ylabel("Number of Subscribers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Finding the average (metrices)
ave_metrices = df[['Subscribers','Visits','Likes','Comments']].mean()/1e3
ave_metrices
```

```python
#Analysing perfomance metrics
plt.figure(figsize=(10,6))
ave_metrices.plot(kind='bar', color=['yellow', 'red', 'black', 'blue'])
plt.title("Average Performance metrics")
plt.xlabel("Metrics")
plt.ylabel("Average")
plt.show()
```

```python
#Checking for patterns using box plot
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['Subscribers','Visits','Likes','Comments']])
plt.title("Perfomance metrice distribution")
plt.xlabel("Metrics")
plt.ylabel("Values")
```

```python
#The analysis of contect categories distribution
cat_distr = df["Categories"].value_counts()
plt.figure(figsize=(10,5))
cat_distr.plot(kind="bar", color="blue")
plt.title("The distribution of content categories")
plt.xlabel("Catergory")
plt.ylabel("Streamers")
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
#Checking the categories with exceptional performance metrics
exceptional_performance = df.groupby("Categories")[['Subscribers','Visits','Likes','Comments']].mean()
plt.figure(figsize=(10,5))
sns.heatmap(exceptional_performance, cmap="viridis", linewidth=.5)
plt.title("The average perfomance metrics by it category")
plt.xlabel("Metrics")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
#Finalizing the analysis with the average perfomance and dispaying the streamers with above-average performance.
ave_subscribers = df["Subscribers"].mean()
ave_visits = df["Visits"].mean()
ave_likes = df["Likes"].mean()
ave_comments = df["Comments"].mean()

ave_streamers = df[
(df["Subscribers"]>ave_subscribers) &
(df["Visits"]>ave_visits) &
(df["Likes"]>ave_likes) &
(df["Comments"]>ave_comments)
 ]
top_performers = ave_streamers.sort_values(by=['Subscribers','Visits','Likes','Comments'], ascending=False) 
print("The top performing content creaters")
print(top_performers[['Username','Subscribers','Visits','Likes','Comments']])
```
